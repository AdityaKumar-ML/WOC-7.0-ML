```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/linear_regression_train.csv")
```

```python
# filling empty values with mean of given values
for i in range((df.shape[1] - 2)):
  df.iloc[:, i + 1] = df.iloc[:, i + 1].fillna(df.iloc[:, i + 1].mean())
```

```python
# using gradient descent
X = np.zeros((df.shape[0], (df.shape[1] - 2)))
for j in range(25):
 X[:,j] = df.iloc[:,j+1]

Y = df['Target'].values

X_b = np.c_[np.ones((X.shape[0],1)), X]

theta = np.random.randn((X.shape[1] + 1), 1)

eta = 0.00001
n_iterations = 5000
m = int(X.shape[0])


for iteration in range(n_iterations):
    gradients = (1/m) * X_b.T.dot(X_b.dot(theta) - Y.reshape(48000,1))
    theta = theta - eta * gradients
```
