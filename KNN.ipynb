```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class KNN():
  def __init__(self, distance_metric):
    self.distance_metric = distance_metric

  def fit(self, X, Y):
    self.X_train = X
    self.Y_train = Y

  def get_distance_metric(self, training_data_point, test_data_point):
    if(self.distance_metric == "eucledian"):

      dist = 0
      for i in range (X.shape[1]):
        dist += (training_data_point[i] - test_data_point[i]) ** 2

      eucledian_dist = np.sqrt(dist)
      return eucledian_dist

    elif (self.distance_metric == "manhattan"):

      dist = 0
      for i in range (X.shape[1]):
        dist += abs(training_data_point[i] - test_data_point[i])

      manhattan_dist = dist
      return manhattan_dist

  def nearest_neighbors(self, X_train, test_data, k):
    distance_list = []
    for training_data in X_train:
      distance = self.get_distance_metric(training_data, test_data)
      distance_list.append((training_data, distance))

    distance_list.sort(key = lambda x: x[1])

    neighbors_list = []

    for j in range (k):
      neighbors_list.append((distance_list[j][0]))

    return neighbors_list

  def predict(self, X_train, test_data, k):
    neighbors = self.nearest_neighbors(X_train, test_data, k)

    for data in neighbors:
      label = []
      label.append(data[-1])

    prediction = max(set(label), key = label.count)
    return prediction
```

```python
df = pd.read_csv("/content/multi_classification_train.csv")
```

```python
X = np.zeros((df.shape[0], df.shape[1] - 2))
for feature in range (df.shape[1] - 2):
  X[:, feature] = df.iloc[:, feature + 1]

Y = df.iloc[:, -1]
```

```python
Classifier = KNN(distance_metric = "eucledian")
Classifier.fit(X, Y)
```

```python
df_test = pd.read_csv("/content/multi_classification_test.csv")

X_test = np.zeros((df_test.shape[0], df_test.shape[1]-1))
for j in range ((df_test.shape[1] - 1)):
    X_test[:,j] = df_test.iloc[:, j+1]
```
