```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class Logistic_Regression():

  # learning_rate = float(input("Enter the learning rate: "))
  # no_of_iterations = int(input("Enter the number of iterations: "))

  learning_rate = 0.00001
  no_of_iterations = 10000


  def _init_(self, learning_rate, no_of_iterations):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations

  def fit(self, X, Y):

    self.m, self.n = X.shape

    self.theta = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    for i in range (self.no_of_iterations):
      self.update_theta()

  def update_theta(self):

    Y_cap = 1 / (1 + np.exp(-(self.X.dot(self.theta) + self.b)))

    d_theta = (1/self.m)*np.dot(self.X.T, (Y_cap - self.Y))
    d_b = (1/self.m)*np.sum(Y_cap - self.Y)

    self.theta = self.theta - self.learning_rate*d_theta
    self.b = self.b - self.learning_rate*d_b

  def predict(self, X):

    Y_pred = 1 / (1 + np.exp(-(X.dot(self.theta) + self.b)))
    Y_pred = np.where(Y_pred > 0.5, 1, 0)
    return Y_pred
```

```python
df_logreg = pd.read_csv("/content/binary_classification_train.csv")
```

```python
X = np.zeros((df_logreg.shape[0], df_logreg.shape[1]-2))
Y = np.zeros((df_logreg.shape[0], 1))

for j in range ((df_logreg.shape[1] - 2)):
  X[:,j] = df_logreg.iloc[:, j+1]

Y = df_logreg.iloc[:, -1]
```

```python
Logreg = Logistic_Regression()
```

```python
Logreg.fit(X, Y)
```

```python
df_test = pd.read_csv("/content/binary_classification_test.csv")
```

```python
X_test = np.zeros((df_test.shape[0], df_test.shape[1]-1))
for j in range ((df_test.shape[1] - 1)):
    X_test[:,j] = df_test.iloc[:, j+1]
```

```python
Y_pred = Logreg.predict(X_test)
```
